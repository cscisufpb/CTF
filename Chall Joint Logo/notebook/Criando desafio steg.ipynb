```python
import cv2 #Importando OpenCV
from matplotlib import pyplot as plt #Importando MatPlotLib para plotar imagem
%matplotlib inline
```

```python
logo = cv2.imread("../img/logo.png") #Carregando as imagens
flag = cv2.imread("../img/flag.png")
```

```python
flag[flag > 0] = 1  #Fazendo flag binaria e sÃ³ com um canal
flag = flag[:,:,0]
```

```python
red = logo[:,:,0] #Pegando a banda onde vai ser colocada a flag

height, width = red.shape

for i in range(height): #Varrendo a banda e setando ultimo bit correpondedo a imagem binaria da flag
    for j in range(width):
        red[i,j] = ((red[i,j] >> 1) << 1) | flag[i,j]

logo[:,:,0] = red #Setando a banda agora com a flag

#cv2.imwrite("../img/challenge.png", logo) #Salvando a imagem com a flag
```

```python
plt.imshow(cv2.cvtColor(logo, cv2.COLOR_BGR2RGB))
```
